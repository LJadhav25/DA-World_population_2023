# How has the population of each country changed from 1970 to 2023? Visualize the trend over time.

```python
que_one = population[['country','1970 population','1980 population','1990 population','2000 population','2010 population','2015 population','2020 population','2022 population','2023 population']]
```

```python
que_one.set_index('country',inplace=True)
```

```python
que_one.head()
```

```python
que_one.transpose().plot(kind='bar', figsize=(10, 6))
plt.xticks(rotation=90)
plt.title('Population Trends of Each Country (1970-2023)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
```

# Which countries have experienced the highest population growth rate between 2000 and 2023? Visualize the top 10 countries.

```python
population.head()
```

```python
que_2 = population[['country','2000 population','2023 population']]
```

```python
que_2.set_index('country',inplace=True)
```

```python
que_2.head()
```

```python
que_2['growth_rate'] = round((que_2['2023 population'] - que_2['2000 population']) / que_2['2000 population']*100,2)
```

```python
top_10_countries = que_2.sort_values(by='growth_rate', ascending=False).head(10)
```

```python
top_10_countries.head()
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_10_countries.index, top_10_countries['growth_rate'], color='skyblue')
plt.xlabel('Growth Rate (%)')
plt.ylabel('Country')
plt.title('Top 10 Countries with Highest Growth Rates (2000-2023)')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest growth rate at the top
plt.grid(axis='x')  # Show grid lines only on the x-axis
```

# What is the population density (people per square kilometer) for each continent? Visualize this using a bar chart or a map.

```python
population.groupby('continent')['continent'].count()
```

```python
population['Population Density'] = round(population['2023 population'] / population['area (km²)'],2)
```

```python
population.head()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(population['continent'], population['Population Density'], color='skyblue')
plt.ylabel('Population Density (people per sq. km)')
plt.xlabel('Continent')
plt.title('Population Density by Continent')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Show grid lines only on the y-axis
plt.tight_layout()
```

# How does the population of each continent compare in terms of percentage of the world population in 2023? Visualize this using a pie chart or a treemap.


```python
population.head()
```

```python
continent_sum = population.groupby('continent')['2023 population'].sum().reset_index(name='Sum continent')
```

```python
continent_sum['Population Percentage Continentwise'] = round((population['2023 population'] / continent_sum['Sum continent']) * 100,2)
```

```python
continent_sum.head()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(continent_sum['Population Percentage Continentwise'], labels=continent_sum['continent'], autopct='%1.1f%%', startangle=140)
plt.title('Population Percentage of Each Continent in 2023')
plt.axis('equal')
```

# How has the population of each country changed from 2010 to 2023? Are there any noticeable trends or outliers? Visualize this using a line chart or a heatmap.


```python
population.head()
```

```python
population_2010_2023 = population[['country','2010 population','2015 population','2020 population','2022 population','2023 population']]
```

```python
population_2010_2023.head()
```

```python
plt.figure(figsize=(12, 8))
population_2010_2023.T.plot(legend=True, marker='o')
plt.title('Population of Each Country from 2010 to 2023')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(population_2010_2023.T, cmap='YlGnBu', linewidths=0.5)
plt.title('Population of Each Country from 2010 to 2023')
plt.xlabel('Country')
plt.ylabel('Year')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# How has the population of the world changed over the past few decades (1990 to 2023)? Visualize this using a stacked area chart or a choropleth map.


```python
population.head()
```

```python
population.info()
```

```python
population_1990_2023 = population[['country','1990 population','2000 population','2010 population','2015 population','2020 population','2022 population','2023 population']]
```

```python
population_1990_2023.set_index('country',inplace=True)
```

```python
population_1990_2023.head()
```

```python
plt.figure(figsize=(12, 8))
population_1990_2023.T.plot(kind='area', stacked=True)
plt.title('World Population Over Time (1990-2023)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
```
